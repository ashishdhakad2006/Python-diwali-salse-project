# Exploratory Data Analysis

# Gender

```python
df.columns
```

```python
ax = sns.countplot(x= 'Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending=False)
```

```python
sales_gen= df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen,palette={"F": "orange", "M": "skyblue"})
```

From above graph we can see that most of the buyers are females and even the purchesing power of the females are grater than men 

 Age

```python
df.columns
```

```python
ax= sns.countplot(data=df, x='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total Amount vs Age group
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount', data= sales_age,palette='tab10')
```

From above graphs we can see that  most of the buyers are age group between 26-35 yers are female

State

```python
df.columns
```

```python
# total numbers of orders from top 10 states
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x= 'State',y='Orders',palette='tab10')
```

```python
#total number of amount/salse in top 10 state
salse_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=salse_amount,x='State',y='Amount',palette='tab10')
```

From the above graph we can see that most of orders and slase/amount are utter pradesh ,maharashtra ,karnataka respectivley

Product_category

```python
sns.set(rc={'figure.figsize':(25,7)})
ax=sns.countplot(data=df,x='Product_Category',palette='tab10')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,7)})
sns.barplot(data= sales_state, x='Product_Category',y='Amount',palette='tab10')
```

From above graphs we can see that the most of the products are from Food,Footwear, Electronic category 

```python
sales_state= df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state, x='Product_ID',y='Orders',palette='tab10')
```
